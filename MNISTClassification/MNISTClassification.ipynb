## 1 hidden layer neural network 



This model simply has the input and output layer with one hidden layer in between with reLU activation. The output layer also has sigmoid activation function because we want probabilities that are between 0 and 1.

```python
class oneHidden(nn.Module) :
    
    def __init__(self, input_size, hlayer_size, output_size) :
        super().__init__()
        self.lin1 = nn.Linear(input_size, hlayer_size, bias=True)
        self.lin2 = nn.Linear(hlayer_size, output_size, bias=True)
        self.relu = nn.ReLU()
        self.sigm = nn.Sigmoid()

    def forward(self, x) :
        x = self.lin1(x)
        x = self.relu(x)
        x = self.lin2(x)
        x = self.sigm(x)
        return x
        
```

```python
model = oneHidden(784, 300, 10).cuda()
```
