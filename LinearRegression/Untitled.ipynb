# Linear Regression using Pytorch 



Generating data 

```python
x = np.linspace(1, 10 ,100)
y = (2 * x) + np.random.randn(100)
```

```python
x = torch.from_numpy(x).type(torch.DoubleTensor)
y = torch.from_numpy(y).type(torch.DoubleTensor)
```

```python
plt.scatter(x, y)
```

```python
a = torch.tensor([-1]).type(torch.DoubleTensor)
b = torch.tensor([1]).type(torch.DoubleTensor)
```

```python
a = torch.nn.Parameter(a)
b = torch.nn.Parameter(b)
```

```python
y_hat = x* a + b
```

```python
plt.scatter(x, y)
plt.scatter(x, y_hat.detach())
```

```python
def mse(y, y_hat) : return ((y-y_hat)**2).mean()
```

```python
def update(c, d, x, y, lr, epochs) :
    losses = []
    for i in range(0, epochs) :
        y_hat = c * x + d
        loss = mse(y, y_hat)
        loss.backward()
    
        with torch.no_grad() :
            c = c.sub_(lr * c.grad)
            d = d.sub_(lr * d.grad)
            c.grad.zero_()
            d.grad.zero_()
        if i%10 == 0 : print(f'loss : {loss}')
        losses.append(loss)
    return losses
```

```python
losses = update(a, b, x, y, 0.01, 100)
```

```python
y_hat = x * a + b
plt.scatter(x, y)
plt.scatter(x, y_hat.detach())
```
